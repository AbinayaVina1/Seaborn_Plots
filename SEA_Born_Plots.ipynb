```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset('diamonds')
print(diamonds.head())
```

```python
sns.scatterplot(data=diamonds, x='carat', y='price')
plt.show()
```

```python
sns.lineplot(data=diamonds.head(100), x='carat', y='price')
plt.show()
```

```python
sns.barplot(data=diamonds, x='cut', y='price')
plt.show()
```

```python
sns.boxplot(data=diamonds, x='cut', y='price')
plt.show()
```

```python
sns.violinplot(data=diamonds, x='cut', y='price')
plt.show()
```

```python
sns.histplot(data=diamonds, x='price', bins=30)
plt.show()
```

```python
sns.kdeplot(data=diamonds['price'], shade=True)
plt.show()
```

```python
sns.countplot(data=diamonds, x='cut')
plt.show()
```

```python
sns.pairplot(diamonds.sample(200), hue='cut')
plt.show()
```

```python
sns.stripplot(data=diamonds.sample(100), x='cut', y='price')
plt.show()
```

```python
sns.swarmplot(data=diamonds.sample(100), x='cut', y='price')
plt.show()
```
